6. Create a notebook parameter: Edit > add_parameter. Name it "my_initials" 

The following code is a SQL block but uses the initials in the notebook parameter. It will drop (delete) the table you created earlier

```python
%sql
DROP TABLE IF EXISTS devcatalog.default.`:my_initials`;
```

The notebook parameter can be used within any cell in the notebook and between languages

```python
#In Python:
my_initials = dbutils.widgets.get("my_initials")
print("you are:",my_initials)
```

```python
%r
#In R:
my_initials <- dbutils.widgets.get("my_initials")
print(paste("you are:",my_initials))
```
